```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_excel("http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx")
```

```python
df
```

```python
countries = df['Country'].unique()
print(countries)
```

```python
df_list_by_country = []
for country in countries:
  df_list_by_country.append(df[df['Country'] == country])
```

```python
df_list_by_country[2]
```

```python
col = 6
print(len(df_list_by_country[0]))
```

```python
# Data preprocessing
records_all_countries = []
record_each_country = []
# I dropped the UK data because it is about half milion rows
# I was running the data including UK and it did not finish for 4 hours...
for country_idx in range(1, len(df_list_by_country)):
  for i in range(len(df_list_by_country[country_idx])):
    record_each_country.append([str(df_list_by_country[country_idx].values[i,j]) for j in range(col)])
  records_all_countries.append(record_each_country)
```

```python
results = []
for country_idx in range(len(records_all_countries)):
  association_rules = apriori(records_all_countries[country_idx], min_support=0.045, min_confidence=0.8)
  association_results = list(association_rules)
  results.append(association_results)
```

```python
def print_association(association_results):
  # reference from lab_10 notebook
  count = 0
  for item in association_results:
    # first index of the inner list
    # Contains base item and add item
    if count == 5:
      break
    count += 1
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("=====================================")
```

```python
record = []
for i in range(len(df_list_by_country[2])):
  record.append([str(df_list_by_country[2].values[i,j]) for j in range(col)])
association_rules = apriori(record, min_support=0.0045, min_confidence=1.0, min_length=3)
association_results = list(association_rules)
```

```python
print("The country is ",countries[2] )
print_association(association_results)
```
