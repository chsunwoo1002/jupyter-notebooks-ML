```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ML/cleanDf.csv')

cleanDf = df.drop(['Votes','UnemploymentRate2015','PovertyAllAgesPct2014','DeepPovAll','AgeTotalPop','MaleRate','FemaleRate','TurnoutRate','Old','Young','Adult',
                   'NomalizePopulation','NomalizeMedHHInc2014','NomalizeToalArea','NomalizePopDensity','State','County'], axis=1)
# I delete the  'State' and 'County' eventhough it is categorical variables
# This is because this variable makes the tree to overfit.
# Same set up as problem 2

headers = cleanDf.columns.tolist()

features = headers.copy()
features.remove('Democrat')
print(headers)
print(features)

cleanDf = pd.get_dummies(cleanDf)
cleanDf
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# convert categorical data into one-hot encoding
X = pd.get_dummies(cleanDf.loc[:, cleanDf.columns != 'Democrat'])
y = cleanDf['Democrat']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)

# 50 tress and set max_depth as 3
# The function max depth is limited to 3, so it is pre-pruning tree
classifier = RandomForestClassifier(n_estimators=50, max_depth=3)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

print(classification_report(y_test, y_pred))
```
