```python
# Set random seed as 1337
import numpy as np
import pandas as pd

np.random.seed(1337)
```

```python

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ML/cleanDf.csv')
```

```python
cleanDf = df.drop(['Votes','UnemploymentRate2015','PovertyAllAgesPct2014','DeepPovAll','AgeTotalPop','MaleRate','FemaleRate','TurnoutRate','Old','Young','Adult',
                   'NomalizePopulation','NomalizeMedHHInc2014','NomalizePerCapitaInc','NomalizeToalArea','NomalizePopDensity','State','County'], axis=1)
# I delete the  'State' and 'County' even though it is categorical variables
# This is because this variable makes the tree to overfit.

```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix


X = pd.get_dummies(cleanDf.loc[:, cleanDf.columns != 'Democrat']) # X is every column except the label vector which is 'Democrat'
y = cleanDf['Democrat'] # Label vector

# Function that suffles all rows and 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, shuffle=True)

classifier = tree.DecisionTreeClassifier(criterion='entropy')
classifier.fit(X_train, y_train) # Training with training data set

y_pred = classifier.predict(X_test)

print(classification_report(y_test, y_pred))
```
