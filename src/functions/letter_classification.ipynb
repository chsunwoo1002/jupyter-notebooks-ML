```python
import numpy as np
import matplotlib.pyplot as plt 

from tensorflow import keras
from tensorflow.keras import layers
from keras.utils.np_utils import to_categorical
```

```python
from google.colab import drive

drive.mount("/content/drive/")
```

```python
from emnist import list_datasets
list_datasets()
```

```python
# get training data set
from emnist import extract_training_samples
train_images, train_labels = extract_training_samples('letters')
```

```python
# get testing data set
from emnist import extract_test_samples
test_images, test_labels = extract_test_samples('letters')
```

```python
# verify the sample sizes
print(train_images.shape)
print(train_labels.shape)
print(test_images.shape)
print(test_labels.shape)
```

```python
train_images = train_images/255
test_images = test_images/255
```

```python
train_images = train_images.reshape(train_images.shape[0], 28, 28,1)
test_images = test_images.reshape(test_images.shape[0], 28, 28,1)
```

```python
train_images.shape
```

```python
np.unique(test_labels)
```

```python
test_labels = to_categorical(test_labels, num_classes=26)
train_labels = to_categorical(train_labels, num_classes=26)
```

```python
model = keras.Sequential(
    [
      layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28, 1)),
      layers.Conv2D(32, (3,3), activation='relu'),
      layers.Conv2D(64, (3,3), activation='relu'),
      layers.Conv2D(64, (3,3), activation='relu'),
      layers.MaxPooling2D(),
      layers.Flatten(), # need to be flatted from 2d to 1d
      layers.Dense(512, activation='relu'),
      layers.Dropout(0.2),
      layers.Dense(27, activation='softmax'),
   ])
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
history = model.fit(x=train_images, y=train_labels, batch_size=256, epochs=10)
```

```python
model.evaluate(x=test_images, y=test_labels)
```
